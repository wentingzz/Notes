```python
def inversions(arr, idx = 0):
    n = len(arr)
    if n == 1:
        return arr, 0, dict()
    else:
        n2 = n//2
        left, l_c, l_i = inversions(arr[:n2], idx )
        right, r_c, r_i = inversions(arr[n2:], idx + n2)
        
        left_n = len(left)
        right_n = len(right)
        sort = [0 for i in range(n)]
        res = {**l_i, **r_i}
        j = 0
        k = 0
        count = 0
        for i in range(n):
            if j == left_n:
                sort[i] = right[k]
                k += 1
            elif k == right_n:
                sort[i] = left[j]
                j += 1
            else:
                if left[j] <= right[k]:
                    sort[i] = left[j]
                    j += 1
                else:
                    sort[i] = right[k]
                    count += left_n - j
                    for p in range(left_n - j):
                        try:
                            res[left[j + p]].append(right[k])
                        except:
                            res[left[j + p]] = [right[k]]
                    k += 1
        return sort, l_c + r_c + count, res

def max_profit(dp, idx):
    res = 0
    for key, values in dp.items():
        if key >= idx:
            for v in values:
                tmp = max_profit(dp, v[0])
                res = max(res, v[1] + tmp)
                print(key, v[1],tmp)
    return res



def maxProfit(prices):
    
    prices = prices[::-1]
    _, count, tmp = inversions(prices)
    print(prices, tmp)
    dic = dict()
    l = len(prices)
    for i in range(l):
        dic[prices[l-i-1]] = l-i-1

    ins = dict()
    for key, values in tmp.items():
        ins[dic[key]] = [[dic[v], key-v] for v in values]
    print(dic, ins)
    try:
        profit = max_profit(ins, min(ins.keys()))
    except:
        profit = 0
    return profit
    
if __name__ == "__main__":
    while(True):
        in_str = input("Input integers separated by comma (ex. 14,15):")
        in_str = in_str.replace(" ", "")
        in_str = in_str.split(",")
        try:
            for i in range(len(in_str)):
                in_str[i] = int(in_str[i])
            break
        except:
            print("The input is not integer or is not separated by comma")
    print(maxProfit(in_str))
#     print("There are", count, "inversions in the given list: ", inversions)[2,1,2,0,1]
```

```python
def merge_sort(arr):
    n = len(arr)
    if n == 1:
        return arr
    else:
        n2 = n//2
        left = merge_sort(arr[:n2])
        right = merge_sort(arr[n2:])
        left_n = len(left)
        right_n = len(right)
        res = ""
        j = 0
        k = 0
        for i in range(n):
            if j == left_n:
                res += right[k]
                k += 1
            elif k == right_n:
                res += left[j]
                j += 1
            else:
                if left[j] < right[k]:
                    res += left[j]
                    j += 1
                else:
                    res += right[k]
                    k += 1
        return res

def groupAnagrams(strs):
    dic = dict()
    for i in range(len(strs)):
        try:
            dic[merge_sort(strs[i])].append(strs[i])
        except:
            dic[merge_sort(strs[i])] = [strs[i]]
    res = []
    for k, vs in dic.items():
        res.append(vs)
    return res
# The following is the solution with better performance (from Leetcode)
#     d = collections.defaultdict(list)
#     for x in strs:
#         xl = ''.join(sorted(x))
#         d[xl].append(x)
#     return d.values()
    

in_str = ["eat", "tea", "tan", "ate", "nat", "bat"]
groupAnagrams(in_str)
# print(merge_sort("eat"))
```
